# Write a program that:
1. Take the file as an input
2. Programmatically analyze the file and print in console the name and position of employees

```python
import pandas as pd

def analyze_excel_file(file_path):
    # Read the Excel file into a pandas DataFrame
    try:
        df = pd.read_excel(file_path)
    except Exception as e:
        print(f"Error reading the Excel file: {e}")
        return

    # Check if required columns are present
    required_columns = ['Employee Name', 'Position ID']
    missing_columns = [col for col in required_columns if col not in df.columns]

    if missing_columns:
        print(f"Error: Required columns are missing - {', '.join(missing_columns)}")
        return

    # Print the name and position of employees
    print("Employee Analysis:")
    for index, row in df.iterrows():
        employee_name = row['Employee Name']
        position_id = row['Position ID']
        print(f"Employee Name: {employee_name}, Position ID: {position_id}")

if __name__ == "__main__":
    # Replace 'your_excel_file.xlsx' with the actual file path
    file_path = '/content/Assignment_Timecard.xlsx'

    analyze_excel_file(file_path)
```

# a) who has worked for 7 consecutive days.

```python
import pandas as pd

def find_employees_with_7_consecutive_days(file_path):
    # Read the Excel file into a pandas DataFrame
    try:
        df = pd.read_excel(file_path)
    except Exception as e:
        print(f"Error reading the Excel file: {e}")
        return

    # Check if required columns are present
    required_columns = ['Time', 'Time Out']
    missing_columns = [col for col in required_columns if col not in df.columns]

    if missing_columns:
        print(f"Error: Required columns are missing - {', '.join(missing_columns)}")
        return

    # Convert 'Time' and 'Time Out' columns to datetime objects
    df['Time'] = pd.to_datetime(df['Time'])
    df['Time Out'] = pd.to_datetime(df['Time Out'])

    # Identify employees who have worked for 7 consecutive days
    consecutive_days_threshold = 7
    consecutive_days_count = 0
    current_employee = None

    for index, row in df.iterrows():
        if current_employee is None or row['Time'].date() == current_employee['Time Out'].date():
            consecutive_days_count += 1
        else:
            consecutive_days_count = 1

        current_employee = row

        if consecutive_days_count == consecutive_days_threshold:
            print(f"Employee who worked for 7 consecutive days: {current_employee['Employee Name']}")
            consecutive_days_count = 0

if __name__ == "__main__":
    # Replace 'your_excel_file.xlsx' with the actual file path
    file_path = '/content/Assignment_Timecard.xlsx'

    find_employees_with_7_consecutive_days(file_path)
```

# b) who have less than 10 hours of time between shifts but greater than 1 hour

```python
import pandas as pd

def find_employees_between_shifts(file_path):
    # Read the Excel file into a pandas DataFrame
    try:
        df = pd.read_excel(file_path)
    except Exception as e:
        print(f"Error reading the Excel file: {e}")
        return

    # Check if required columns are present
    required_columns = [
        'Position ID', 'Position Status', 'Time', 'Time Out',
        'Timecard Hours (as Time)', 'Pay Cycle Start Date',
        'Pay Cycle End Date', 'Employee Name', 'File Number'
    ]
    missing_columns = [col for col in required_columns if col not in df.columns]

    if missing_columns:
        print(f"Error: Required columns are missing - {', '.join(missing_columns)}")
        return

    # Convert 'Time' and 'Time Out' columns to datetime objects
    df['Time'] = pd.to_datetime(df['Time'])
    df['Time Out'] = pd.to_datetime(df['Time Out'])

    # Identify employees with less than 10 hours but greater than 1 hour between shifts
    for i in range(1, len(df)):
        time_diff = df.loc[i, 'Time'] - df.loc[i - 1, 'Time Out']
        hours_between_shifts = time_diff.total_seconds() / 3600

        if 1 < hours_between_shifts < 10:
            employee_name = df.loc[i, 'Employee Name']
            position_id = df.loc[i, 'Position ID']
            position_status = df.loc[i, 'Position Status']

            print(f"Employee Name: {employee_name}, Position ID: {position_id},")

if __name__ == "__main__":
    # Replace 'your_excel_file.xlsx' with the actual file path
    file_path = '/content/Assignment_Timecard.xlsx'

    find_employees_between_shifts(file_path)
```

# c) Who has worked for more than 14 hours in a single

```python
import pandas as pd

def find_employees_with_long_shifts(file_path):
    # Read the Excel file into a pandas DataFrame
    try:
        df = pd.read_excel(file_path)
    except Exception as e:
        print(f"Error reading the Excel file: {e}")
        return

    # Check if required columns are present
    required_columns = ['Position ID', 'Position Status', 'Time', 'Time Out', 'Timecard Hours (as Time)',
                         'Pay Cycle Start Date', 'Pay Cycle End Date', 'Employee Name', 'File Number']
    missing_columns = [col for col in required_columns if col not in df.columns]

    if missing_columns:
        print(f"Error: Required columns are missing - {', '.join(missing_columns)}")
        return

    # Convert 'Time' and 'Time Out' columns to datetime objects
    df['Time'] = pd.to_datetime(df['Time'])
    df['Time Out'] = pd.to_datetime(df['Time Out'])

    # Filter employees who have worked for more than 14 hours in a single shift
    long_shift_threshold = 14
    for index, row in df.iterrows():
        shift_duration = (row['Time Out'] - row['Time']).total_seconds() / 3600  # Convert seconds to hours
        if shift_duration > long_shift_threshold:
            print(f"Employee with long shift: Employee Name : {row['Employee Name']}, Position ID: {row['Position ID']}, "
                  f"Shift Duration: {shift_duration:.2f} hours")

if __name__ == "__main__":
    # Replace 'your_excel_file.xlsx' with the actual file path
    file_path = '/content/Assignment_Timecard.xlsx'

    find_employees_with_long_shifts(file_path)
```
